```python
import pandas as pd
```

```python

def readfile(path,filename):
    df = pd.read_csv(path+filename,header=0,sep='\t',skipinitialspace=True)
    return df
```

```python
ref_path ='/home/alva/Documents/lncRNAs_project_2020/reference/'
ALL_path ='/home/alva/Documents/lncRNAs_project_2020/mypaper/'
```

```python
ref = readfile(ref_path,"gencode.v19.annotation_gene.tsv")
ALL = readfile(ALL_path,"expression_ALL")
ref['Geneid'] =ref['Geneid'].apply(lambda x:x.split('.')[0])
```

```python
ALL_ref =pd.merge(ref[["Geneid","chr","start","end"]],ALL,on='Geneid')
ALL_ref.to_csv('/home/alva/Documents/lncRNAs_project_2020/mypaper/expression_ALL_withchr', sep='\t',index=False)
```
